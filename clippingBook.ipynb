```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

fig=plt.figure()
ax=Axes3D(fig)

X=np.arange(-4,4,0.25)
Y=np.arange(-4,4,0.25)
X,Y=np.meshgrid(X,Y)
R=np.sqrt(X ** 2 + Y ** 2)
Z=np.sin(R)

ax.plot_surface(X,Y,Z,rstride=1,cstride=1,cmap=plt.get_cmap('rainbow'))

ax.contourf(X,Y,Z,zdir='z',offset=-2,cmap=plt.get_cmap('rainbow'))

plt.show()
```

```python
# iCode, inner product of two vectors
### CLASSIC DOT PRODUCT OF VECTORS IMPLEMENTATION ###
def innerProduct_vectors(x1, x2):
    """
    Implement a function that caculate inner product of two vectors.
    
    Argument:
    x1 -- vector 1
    x2 -- vector 2
    
    Returns:
    dot -- Inner product of the two vectors.
    """
    dot = 0
    for i in range(len(x1)):
        dot+= x1[i]*x2[i]

    return dot
```

```python
# iCode, outer product of two vectors
### CLASSIC outer PRODUCT OF VECTORS IMPLEMENTATION ###
def outerProduct_vectors(x1, x2):
    """
    Implement a function that caculate outer product of two vectors.
    
    Argument:
    x1 -- vector 1
    x2 -- vector 2
    
    Returns:
    outer -- outer product of the two vectors.
    """
    outer = np.zeros((len(x1),len(x2))) # we create a len(x1)*len(x2) matrix with only zeros
    for i in range(len(x1)):
        for j in range(len(x2)):
            outer[i,j] = x1[i]*x2[j]

    return outer
```

```python
# iCode, elementwise product of two vectors
### CLASSIC outer PRODUCT OF VECTORS IMPLEMENTATION ###
def elementwiseProduct_vectors(x1, x2):
    """
    Implement a function that caculate elementwise product of two vectors.
    
    Argument:
    x1 -- vector 1
    x2 -- vector 2
    
    Returns:
    outer -- outer product of the two vectors.
    """
    mul = np.zeros(len(x1))
    for i in range(len(x1)):
            mul[i] = x1[i]*x2[i]

    return mul
```

```python
# iCode, general dot product of two vectors
### CLASSIC GENERAL DOT PRODUCT IMPLEMENTATION ###
def generalDotProduct(x1,x2):
    """
    Implement a function that caculate elementwise product of two vectors.
    
    Argument:
    x1 -- matrix 1
    x2 -- vector 2
    
    Returns:
    outer -- outer product of the two vectors.
    """
    np.random.seed(1)
    gdot = np.zeros(len(x1))
    for i in range(len(x1)):
        for j in range(len(x2)):
            gdot[i] += x1[i,j]*x2[j]
    print("general Dot = " + str(gdot))

    return gdot
```

```python
#iCode, compare performance of np.dot() and **2
import time
np.random.seed(1)
iterations = 10000
y = np.random.randn(10)
yhat = y + np.random.randn()*0.05
print("y =" + str(y) )
print("yhat =" + str(yhat))

#calcuated by **2
tic = time.process_time()
for i in range(iterations):
    loss = np.sum((y - yhat) ** 2)
toc = time.process_time()
print("calcuated by **2 ")
print ("loss = " + str(loss) + "\n ----- Computation time = " + str(1000*(toc - tic)) + "ms")

#calcuated by np.dot
tic = time.process_time()
for i in range(iterations):
    loss = np.dot(y - yhat,y - yhat)
toc = time.process_time()
print("calcuated by np.dot ")
print ("loss = " + str(loss) + "\n ----- Computation time = " + str(1000*(toc - tic)) + "ms")
```

```python
#Tanh函数及其导数
import math
import matplotlib.pyplot as plt
import numpy as np
import mpl_toolkits.axisartist as axisartist
 
# Tanh 激活函数
class Tanh:       
    # 原函数 
    def forward(self, x):
        return (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))
    # 导数
    def backward(self, outx):
        tanh = (np.exp(outx) - np.exp(-outx)) / (np.exp(outx) + np.exp(-outx))
        return 1 - math.pow(tanh, 2)
 
# 画图
def Axis(fig, ax):
    #将绘图区对象添加到画布中
    fig.add_axes(ax)
    # 隐藏坐标抽
    ax.axis[:].set_visible(False)
    # new_floating_axis 创建新的坐标
    ax.axis["x"] = ax.new_floating_axis(0, 0)
    # 给 x 轴创建箭头线，大小为1.0
    ax.axis["x"].set_axisline_style("->", size = 1.0)
    # 给 x 轴箭头指向方向
    ax.axis["x"].set_axis_direction("top")
    # 同理，创建 y 轴
    ax.axis["y"] = ax.new_floating_axis(1, 0)
    ax.axis["y"].set_axisline_style("->", size = 1.0)
    ax.axis["y"].set_axis_direction("right")
 
# 返回间隔均匀的100个样本，计算间隔为[start, stop]。
x =  np.linspace(-10, 10, 100)
y_forward = []
y_backward = []
 
def get_list_forward(x):
    for i in range(len(x)):
        y_forward.append(Tanh().forward(x[i]))
    return y_forward
 
def get_list_backward(x):
    for i in range(len(x)):
        y_backward.append(Tanh().backward(x[i]))
    return y_backward
    
y_forward = get_list_forward(x)
y_backward = get_list_backward(x)
 
#创建画布
fig = plt.figure(figsize=(12, 12))
 
#创建绘图对象ax
ax = axisartist.Subplot(fig, 111)
Axis(fig, ax)
 
# 设置x, y轴范围
plt.ylim((-2, 2))
plt.xlim((-10, 10))
 
# 原函数，forward function
plt.plot(x, y_forward, color='red', label='$f(x) = tanh(x)$')
plt.legend()
 
# 导数， backward function
plt.plot(x, y_backward, label='f(x)\' = 1-(tanh)^2')
plt.legend()
 
plt.show()
```
